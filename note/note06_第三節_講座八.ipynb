---
利用我們學過挑選欄位的方法，各自選出欄位操作一遍  
大略上可以想成  
iloc: 利用index的號碼個數(解釋起來有點怪)，列號/行號。  
loc: 利用標籤、名稱、index的號碼本體。

```python
# 給大家一個粗略的比較
titanic.head()
```

```python
titanic.iloc[:3]
```

```python
titanic.loc[:3]
```

---
以下正式使用groupby功能。

```python
titanic[["Sex", "Age"]].groupby(by = "Sex").mean()
```

```python
titanic.iloc[:, [3,4]].groupby(by = "Sex").mean()
```

```python
titanic.loc[:, ["Sex", "Age"]].groupby(by = "Sex").mean()
```
