## 從大排序到小的前幾名，即最大或最小的N個元素

- nlargest()
- nsmallest()

```python
titanic.sort_values(by = "Fare", ascending = False).head(10)
```

```python
titanic.nlargest(n = 10, columns = "Fare")
# 可以注意看和上面的順序不一樣!!這就是keep參數的用途~
# 
# 有一個keep參數
# keep='first'或者'last'。當出現重複值的時候，keep='first',會選取在原始DataFrame裡排在前面的，keep='last'則去排後面的。
```

```python
titanic.nsmallest(n = 20, columns = "Fare")
# 可以看到有些人是沒有付出任何金錢的。
```

### line plot(預設)

如果直接畫圖，就只會呈現數值類別的資料視覺。

```python
titanic.plot()
plt.show() #最後一行一定要加這個!
# 因為我有PassengerId，所以看起來和影片中不一樣~
```

```python
tmp01 = titanic.drop("PassengerId", axis=1)
# axis : {0 or 'index', 1 or 'columns'}, default 0
```

```python
tmp01.plot()
plt.show() #最後一行一定要加這個!
# 刪去PassenderId就可以跟影片一樣了
```

直接視覺化幾乎看不出任何東西  
因為很rough，因此我們要使用plot()中的參數 subplot

```python
tmp01.plot(subplots = True)
plt.show() #最後一行一定要加這個!
```

但是這樣還是幾乎看不出東西，因為圖片太小~  
所以在使用plot()中的參數 figsize

```python
tmp01.plot(subplots = True, figsize = (12, 15)) #(寬, 高)
plt.show() #最後一行一定要加這個!
```

從這裡就可以很快地看出一些離群值  
例如：小孩很多的、或是票價付很多的。  
像是Fare就有三個人付超過400元，大部分是聚集在0~150之間

---
當然可以只針對某個欄位畫圖。  
參數subplots可以去除，因為只有一個欄位

```python
tmp01.Age.plot(figsize = (12, 8)) #(寬, 高)
plt.show() #最後一行一定要加這個!
```

可以簡單看出Age大多在0~80之間，然後出現最多的是在20~30之間。

---
除了簡單的line plt之外，當然還有Bar plot, scatter plot, Histogram, ...and so on.  
請利用plot()參數的kind做不同的繪圖。

### Histogram

```python
tmp01.Age.plot(kind = "hist", figsize = (12, 8)) #(寬, 高)
plt.show() #最後一行一定要加這個!
```

可以觀察到電腦很聰明，如果是畫histogram，會大約分成10個區間。  
當然我們可以用plot()中的參數bins做調整。  
- bins = 要分成幾個區間

```python
tmp01.Age.plot(kind = "hist", figsize = (12, 8), bins = 80) #(寬, 高)
plt.show() #最後一行一定要加這個!
```

可以看到說有一個人是80歲，但是沒有人是78, 79這樣  
然後最多人的應該是24~26歲區間

```python
# 可以用valur_counts()去檢查，但是注意到Histogram會把24.5計算在24~25區間。
tmp01.Age.value_counts()
```

---
### scatter plot
- 需要設定x-axis, y-axis

```python
tmp01.plot(kind = "scatter", x = "Age", y = "Fare", figsize = (14,8)) #(寬, 高)
plt.xlabel("AGE", fontsize = 20)
plt.ylabel("Ticket Price", fontsize = 20)
plt.title("Titanic Passengers (Age vs. Fare)", fontsize = 25)
plt.show() #最後一行一定要加這個!
```
